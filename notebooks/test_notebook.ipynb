```python
def return_valid_address(address):
    matched_address = re.findall("^(\d+) ?([A-Za-z](?= ))? (.*?) ([^ ]+?) ?((?<= )APT)? ?((?<= )\d*)?$", str(address))
    if len(matched_address) > 0:
        single_string = " ".join(matched_address[0]).strip()
        return ' '.join(single_string.split())
```
